```python
import pandas as pd
import numpy as np
#creating a data set with 4  column along with the dates and indexes starting from 10/10/2005 which is prepared for 10 unique periods
df = pd.DataFrame(np.random.randn(10, 4),
   index = pd.date_range('10/10/2005', periods=10),
   columns = ['I', 'II', 'III', 'iv'])
df
```

```python
#uses rolling function to carryout overall calculation
r = df.rolling(window=3,min_periods=1)
tempSum=r.aggregate(np.sum)
#summing the data present in the form of series
tempSum
```

```python
#selection of 1st column prepared after conducting aggregate sum
r['I'].aggregate(np.sum)
```

```python
#selection of two columns along with dates as indexes 
r[['I','II']].aggregate(np.sum)
```

```python
#presenting the data  selected from column I by carrying out numpy sum and mean 
r['I'].aggregate([np.sum,np.mean])
```

```python
#presenting the data  selected from column I  and II by carrying out numpy sum and mean 
r[['I','II']].aggregate([np.sum,np.mean])
```

```python
#presenting the data of column I by performing numpy sum and column II by performing numpy mean
r.aggregate({'I' : np.sum,'II' : np.mean})
```
